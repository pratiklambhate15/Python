**To remove the outliers I am using IQR method as the data is not normally

```python
q1 = np.quantile(bd['age'], 0.25) # asking for q1 having 0.25 percentile
q2 = np.quantile(bd['age'], 0.50) #asking for q2 having 0.50 percentile
q3 = np.quantile(bd['age'], 0.75) #asking for q3 having 0.75 percentile
print(f'Q1 = {q1} or Q2 = {q2} or Q3 = {q3}')  
```

```python
IQR = q3-q1
```

```python
#Calculating the upper and the lower limit 
upp_limit = q3 + 1.5*IQR 
low_limit = q1 - 1.5*IQR
```

```python
bd[bd['age'] >  upp_limit]['age'] # there are 487 upper values in the age column 
```

```python
bd['age'] = np.where(bd['age'] > upp_limit, upp_limit, bd['age'])
```

# Feature Scaling

```python
X=cpd.iloc[:,0:8]                                   # Applying syntax to X and Y for ILOC Transformation for Feature Scaling
Y=cpd.iloc[:,8]
```

```python
from sklearn.preprocessing import StandardScaler    # Importing Standanrd Scaler for Scaling purpose
sc = StandardScaler()                               # Assigning sc as Standard Scaler
X = sc.fit_transform(X)                             # Applying transformation to X
```

```python
X
```

```python
X_min_max = (X - X.min()) / (X.max() - X.min())           # Applying min max scaling.
X_min_max
```

# Feature Extraction by using PCA

```python
from sklearn.decomposition import PCA             # Importing PCA (Principal Component Analysis) library
pca=PCA()                                         # Assigning pca as PCA
fit=pca.fit(X)                                    # Applying fit to ILOC transformation data
```

```python
PC_values = np.arange(pca.n_components_) + 1      # Applying Scree Plot of a Elbow Plot for finding the kink in data
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot of a Elbow Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)   # Applying Bar Plot to the above Scree Plot 
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

```python
pca.explained_variance_ratio_                  # It shows the exact varience of all Principal Components
```

```python
x_data = pca.transform(X)                      #Transforming x_train
x_data
```

```python
import statsmodels.api as sm                # Importing Stats Model as sm
sm.add_constant(x_data)
```
